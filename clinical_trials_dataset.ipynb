```python
#create age group col
age_bins = [0, 30, 40, 50, 60, 70, 80]
age_labels = ['<30', '30-39', '40-49', '50-59', '60-69', '70+']
df['Age_Group'] = pd.cut(df['age'], bins=age_bins, labels=age_labels, right=False)

print(df[['age', 'Age_Group']].head(10));
```

```python
# Define a function to calculate relative tumor size reduction
def calculate_relative_reduction(row):
    baseline_size = row['MRI_LD_Baseline']
    post_chemo_size = row['MRI_LD_1_3dAC']
    if baseline_size > 0:
        return (baseline_size - post_chemo_size) / baseline_size
    else:
        return 0
```

```python

# Apply the function to create a new column for the full dataset
df['Relative_Tumor_Size_Reduction'] = df.apply(calculate_relative_reduction, axis=1)

print(df[['MRI_LD_Baseline', 'MRI_LD_1_3dAC', 'Relative_Tumor_Size_Reduction']].head())
```

Meaning: while there is a weak positive correlation between baseline tumor size (MRI_LD_Baseline) and relative tumor size reduction after treatment (Relative_Tumor_Size_Reduction), we cant rely solely on baseline tumor size to predict treatment outcomes for this study.

```python
# Calculate survival and death rates
survival_rate = df['Alive'].value_counts(normalize=True)['Yes'] * 100
death_rate = df['Alive'].value_counts(normalize=True)['No'] * 100

print(f"Survival Rate: {survival_rate:.2f}%")
print(f"Death Rate: {death_rate:.2f}%")
```

```python
# Prepare data for bar chart
survival_data = pd.DataFrame({
    'Status': ['Survived', 'Died'],
    'Rate': [survival_rate, death_rate]
})

# Plot bar chart
plt.figure(figsize=(10, 5))
sns.barplot(x='Status', y='Rate', data=survival_data)
plt.title('Survival and Death Rates')
plt.ylabel('Rate (%)')
plt.show();
```
